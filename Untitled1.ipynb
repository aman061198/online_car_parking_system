```python
from tkinter import *
from sqlite3 import *
from tkinter import messagebox
from datetime import *
'''
GUI of front page
'''
global pay_id
pay_id = 100000
file = open("pay_id",'r')
x = int(file.read())
pay_id = x
file.close()
Database = connect('DataBase.db')
Database.execute("create table if not exists user(reg_no number(8) primary key,password varchar(40) not null,\
name varchar(30) not null,DOB char(10),phone_no number(10),vehicle_no varchar(25) not null)")
Database.commit()

Database.execute("create table if not exists parking_slot(slot varchar(5) primary key,\
reg_no number(8) references user(reg_no),status varchar(3) default 'no')")
l = [('A1',None,'no'),('A2',None,'no'),('A3',None,'no'),('A4',None,'no'),('A5',None,'no'),('A6',None,'no')
     ,('A7',None,'no'),('A8',None,'no'),('A9',None,'no'),('A10',None,'no')]
#Database.executemany("insert into parking_slot values(?,?,?)",l)
Database.commit()

Database.execute("create table if not exists payment(payment_id number(6) primary key,reg_no number(8) references user(reg_no),\
amount number(5),paid_on varchar(30),status varchar(10))")
Database.commit()

Database.execute("create table if not exists record(reg_no number(8) references user(reg_no),time_in varchar(30),\
time_out varchar(30))")
Database.commit()
Database.close()

    
def MainPage():
    global root,screen_width,screen_height
    root = Tk()
    screen_width = root.winfo_screenwidth()
    screen_height = root.winfo_screenheight()
    root.geometry("%dx%d" % (screen_width,screen_height))
    root.config(bg="yellow")

    logo = PhotoImage(file="LPU-logo.png")
    logo_image = Label(root,image=logo,bg='yellow')

    root.iconphoto(root, PhotoImage(file="LPU-logo.png")) 
    root.title("PARKING MANAGEMENT SYSTEM")
    
    global x
    Database = connect("DataBase.db")
    x = Database.execute("select count(slot) from parking_slot where status='no'")
    x = x.fetchall()
    x = x[0][0]
    Database.close()
    
    heading = Label(root,text="LPU  CAR  PARKING  MAMAGEMENT  SYSTEM",font=('times', 35, 'italic bold underline'),bg=('yellow')) 
    park = Button(root,text="PARK MY CAR",bd=3,font=('times',25,'bold'),command=ParkMyCar,height=1,width=12)
    call = Button(root,text = "CALL MY CAR",bd=3,font=('times',25,'bold'),command=CallMyCar,height=1,width=12)
    admin = Button(root,text = 'ADMIN login',bd=3,font=('times',25,'bold'),command=AdminLogin,height=1,width=12)
    new_user = Button(root,text = 'NEW USER',bd=3,font=('times',25,'bold'),command=NewUser,height=1,width=12)
    #user_help = Button(root,text = 'HELP',bd=3,font=('times',25,'bold'),height=1,width=12)
    avail = Label(root,text="Number of free slots",font=('times', 25, 'italic bold'),bg=('yellow'))
    value = Label(root,text=x,font=('times', 35, 'italic bold underline'))
    
    logo_image.grid(padx=0)
    heading.grid(row=0,column=1)
    avail.grid(row=1,column=0)
    value.grid(row=2,column=0)
    #user_help.grid(row=5,column = 1,pady=10)
    new_user.grid(row=4,column=1,pady=10)
    admin.grid(row=3,column=1,pady=10)
    call.grid(row=2,column=1,pady=10)
    park.grid(row=1,column=1,pady=10)
    root.mainloop()

def NewUser():
    root.destroy()
    global NewUser_gui
    NewUser_gui = Tk()
    
    NewUser_gui.geometry("%dx%d" % (screen_width,screen_height))
    NewUser_gui.config(bg="yellow")
    
    logo = PhotoImage(file="LPU-logo.png")

    NewUser_gui.iconphoto(NewUser_gui, PhotoImage(file="LPU-logo.png")) 
    NewUser_gui.title("PARKING MANAGEMENT SYSTEM")
      
    heading = Label(NewUser_gui,text="ENTER YOUR DETAILS",font=('times', 35, 'italic bold'),bg='yellow')
    user_id = Label(NewUser_gui,text="Registration Number",font=('times',20,'bold'),bg='yellow')
    name = Label(NewUser_gui,text="Name",font=('times',20,'bold'),bg='yellow')
    dob = Label(NewUser_gui,text="Date Of Birth (yyyy/mm/dd)",font=('times',20,'bold'),bg='yellow')
    phone = Label(NewUser_gui,text="Phone Number",font=('times',20,'bold'),bg='yellow')
    vehicle_no = Label(NewUser_gui,text="vehicle number",font=('times',20,'bold'),bg='yellow')
    password = Label(NewUser_gui,text="Password",font=('times',20,'bold'),bg='yellow')
    confirm = Label(NewUser_gui,text="confirm password",font=('times',20,'bold'),bg='yellow')
    
    global user_id_e,name_e,vehicle_no_e,password_e,confirm_e,dob_e,phone_e
    user_id_e = Entry(NewUser_gui,bd=3,width=50)
    name_e = Entry(NewUser_gui,bd=3,width=50)
    dob_e = Entry(NewUser_gui,bd=3,width=50)
    phone_e = Entry(NewUser_gui,bd=3,width=50)
    vehicle_no_e = Entry(NewUser_gui,bd=3,width=50)
    password_e = Entry(NewUser_gui,bd=3,width=50,show='*')
    confirm_e = Entry(NewUser_gui,bd=3,width=50,show='*')
    
    create = Button(NewUser_gui,text="Create",font=('times',20,'bold'),command=CreateUser)
    back = Button(NewUser_gui,text="Back",font=('times',20,'bold'),command=lambda: GoBack(NewUser_gui))
    
    heading.grid(padx=150)
    user_id.grid(row=1,column=0,pady=20)
    name.grid(row=2,column=0,pady=20)
    dob.grid(row=3,column=0,pady=20)
    phone.grid(row=4,column=0,pady=20)
    vehicle_no.grid(row=5,column=0,pady=20)
    password.grid(row=6,column=0,pady=20)
    confirm.grid(row=7,column=0,pady=20)
    create.grid(row=8,column=0,pady=40)
    back.grid(row=8,column=1,pady=40)
    
    user_id_e.grid(row=1,column=1,pady=20)
    name_e.grid(row=2,column=1,pady=20)
    dob_e.grid(row=3,column=1,pady=20)
    phone_e.grid(row=4,column=1,pady=20)
    vehicle_no_e.grid(row=5,column=1,pady=20)
    password_e.grid(row=6,column=1,pady=20)
    confirm_e.grid(row=7,column=1,pady=20)
    
    NewUser_gui.mainloop()
    
    
def CreateUser():
    reg_no = user_id_e.get()
    name = name_e.get()
    dob = dob_e.get()
    phone = phone_e.get()
    vehicle_no = vehicle_no_e.get()
    password = password_e.get()
    confirm = confirm_e.get()
    #----------------------VALIDATION and UPDATION------------------------#
    if len(reg_no)!=8  or (not reg_no.isdigit()):
        messagebox.showinfo("registraion number","invalid registration number \n registraion number should only contain numbers of length 8")
    else:
        if dob[4] != '/' or dob[7] != '/' or not(dob[:4].isdigit() and dob[5:7].isdigit() and dob[8:10].isdigit()):
            messagebox.showinfo("wrong details","invalid date of birth formate")
        elif len(phone) != 10 or (not phone.isdigit()):
            messagebox.showinfo("wrong details","invalid phone number")
        elif password != confirm :
            messagebox.showinfo("password","password doesnot match with confirm password")    
        else:
            ask = messagebox.askquestion("Confirmation","Are you sure want to continue")
            if ask=="yes":
                Database = connect("DataBase.db")
                s = Database.execute("select reg_no from user")
                Database.commit()
                #Database.close()
                if reg_no in str(s.fetchall()):
                    messagebox.showinfo("user exist","user already exist so try to login")
                    #Database = connect("DataBase.db")
                else:
                    reg_no = int(reg_no)
                    phone = int(phone)
                    Database.execute("insert into user values(?,?,?,?,?,?)",(reg_no,password,name,dob,phone,vehicle_no))
                    Database.commit()
                    messagebox.showinfo("login sucessfull","use your user_id and password for login")
                    NewUser_gui.destroy()
                    Database.close()
                    MainPage()
                    
            else:
                pass
            
                
def ParkMyCar():
    if  x>0:
        root.destroy()
        global ParkMyCar_gui
        ParkMyCar_gui = Tk()
    
        ParkMyCar_gui.geometry("%dx%d" % (screen_width,screen_height))
        ParkMyCar_gui.config(bg="yellow")
    
        logo = PhotoImage(file="LPU-logo.png")
        logo_image = Label(ParkMyCar_gui,image=logo,bg='yellow')

        ParkMyCar_gui.iconphoto(ParkMyCar_gui, PhotoImage(file="LPU-logo.png")) 
        ParkMyCar_gui.title("PARKING MANAGEMENT SYSTEM")
    
        heading = Label(ParkMyCar_gui,text="LPU  CAR  PARKING  MAMAGEMENT  SYSTEM",font=('times', 35, 'italic bold underline'),bg=('yellow')) 
    
        user_id = Label(ParkMyCar_gui,text="User ID",font=('times',20,'bold'),bg='yellow')
        password = Label(ParkMyCar_gui,text="Password",font=('times',20,'bold'),bg='yellow')
    
        global user_id_e,password_e
        user_id_e = Entry(ParkMyCar_gui,bd=3,width=50)
        password_e = Entry(ParkMyCar_gui,bd=3,width=50)
    
        confirm = Button(ParkMyCar_gui,text="Confirm",font=('times',20,'bold'),command=Login)
        back = Button(ParkMyCar_gui,text="Back",font=('times',20,'bold'),command = lambda: GoBack(ParkMyCar_gui))

        logo_image.grid(row=0,column=0)
        heading.grid(row=0,column=1)
        user_id.grid(row=2,column=0,pady=30)
        password.grid(row=4,column=0,pady=30)
        user_id_e.grid(row=2,column=1,pady=30)
        password_e.grid(row=4,column=1,pady=30)
        confirm.grid(row=6,column=0,pady=30,padx=80)
        back.grid(row=6,column=1,pady=30,padx=30)
    
        ParkMyCar_gui.mainloop()
    else:
        messagebox.showinfo("slots","No slots are empty \n Sorry for inconvinience")
def GoBack(gui):
    gui.destroy()
    MainPage()    


def Login():
    user_id=user_id_e.get()
    password=password_e.get()
    global user
    user = (int(user_id),str(password))
    
    Database = connect("DataBase.db")
    s = Database.execute("select reg_no,password from user")
    #Database.close()
    s = s.fetchall()
    if user in s:
        ask=messagebox.askquestion("confirmation","Are you sure want to park")
        if ask=="yes":
            Database.close()
            ParkingConfirm()
        else:
            pass
    else:
        messagebox.showerror("error","user id password do not exist")

def ParkingConfirm():
    Database = connect("DataBase.db")
    record = Database.execute("select reg_no from record where time_out='-'")
    record = record.fetchall()
    print(record)
    print(user[0])
    if (user[0],) in record:
        messagebox.showerror("alert","your one vehicle is alredy in there\n so take it out first")
    
    else:
        s = Database.execute("select * from parking_slot")
        s = s.fetchall()
        Database.close()
        for i in s:
            if i[2]=='no':
                e_slot = i[0]
                break
        Database = connect("DataBase.db")
        temp = user[0]
        Database.execute("update parking_slot set reg_no = (?),status='Yes' where slot = (?)",(temp,e_slot))
        Database.commit()
        now = str(datetime.now())
        temp = user[0]
        Database.execute("insert into record values(?,?,?)",(temp,now,'-'))
        Database.commit()
        messagebox.showinfo("sucessfull","your vehicle is parked in "+e_slot)
        ParkMyCar_gui.destroy()
        MainPage()

        
def CallMyCar():
    if  x < 10:
        root.destroy()
        global CallMyCar_gui
        CallMyCar_gui = Tk()
    
        CallMyCar_gui.geometry("%dx%d" % (screen_width,screen_height))
        CallMyCar_gui.config(bg="yellow")
    
        logo = PhotoImage(file="LPU-logo.png")
        logo_image = Label(CallMyCar_gui,image=logo,bg='yellow')

        CallMyCar_gui.iconphoto(CallMyCar_gui, PhotoImage(file="LPU-logo.png")) 
        CallMyCar_gui.title("PARKING MANAGEMENT SYSTEM")
    
        heading = Label(CallMyCar_gui,text="LPU  CAR  PARKING  MAMAGEMENT  SYSTEM",font=('times', 35, 'italic bold underline'),bg=('yellow')) 
    
        user_id = Label(CallMyCar_gui,text="User ID",font=('times',20,'bold'),bg='yellow')
        password = Label(CallMyCar_gui,text="Password",font=('times',20,'bold'),bg='yellow')
    
        #global user_id_e,password_e
        user_id_en = Entry(CallMyCar_gui,bd=3,width=50)
        password_en = Entry(CallMyCar_gui,bd=3,width=50)
    
        confirm = Button(CallMyCar_gui,text="Confirm",font=('times',20,'bold'),command=lambda: LoginCall(user_id_en,password_en))
        back = Button(CallMyCar_gui,text="Back",font=('times',20,'bold'),command=lambda: GoBack(CallMyCar_gui))

        logo_image.grid(row=0,column=0)
        heading.grid(row=0,column=1)
        user_id.grid(row=2,column=0,pady=30)
        password.grid(row=4,column=0,pady=30)
        user_id_en.grid(row=2,column=1,pady=30)
        password_en.grid(row=4,column=1,pady=30)
        confirm.grid(row=6,column=0,pady=30,padx=80)
        back.grid(row=6,column=1,pady=30,padx=30)
    
        CallMyCar_gui.mainloop()
    else:
        messagebox.showinfo("slots","No car is in there")

    
def LoginCall(user_id_en,password_en):
    user_id=user_id_en.get()
    password=password_en.get()
    global user
    user = (int(user_id),str(password))
    Database = connect("DataBase.db")
    s = Database.execute("select reg_no,password from user")
    #Database.close()
    s = s.fetchall()
    if user in s:
        ask=messagebox.askquestion("confirmation","Are you sure want to call your vehicle")
        if ask=="yes":
            Database.close()
            CallConfirm()
        else:
            pass
    else:
        messagebox.showerror("error","user id password do not exist")

def CallConfirm():
    temp = user[0]
    Database = connect("DataBase.db")
    park_time = Database.execute("select time_in from record where reg_no=(?) and time_out=(?)",(temp,'-'))
    park_time = park_time.fetchone()[0]
    park_time = datetime.strptime(park_time,'%Y-%m-%d %H:%M:%S.%f')
    current_time = datetime.now()
    t = str(current_time)
    
    Database.execute("update record set time_out = (?) where reg_no=(?)",(t,temp))
    Database.commit()
    Database.execute("update parking_slot set status='no',reg_no='None' where reg_no=(?)",(temp,))
    Database.commit()
    
    hours_used = (current_time - park_time).total_seconds()/3600
    Database.close()
    CallMyCar_gui.destroy()
    Payment(hours_used)
    
def Payment(hours_used):    
    global amount,Payment_gui
    amount = hours_used*10
    Payment_gui = Tk()
    Payment_gui.geometry("%dx%d" % (screen_width,screen_height))
    Payment_gui.config(bg="yellow")
    
    logo = PhotoImage(file="LPU-logo.png")
    logo_image = Label(Payment_gui,image=logo,bg='yellow')

    Payment_gui.iconphoto(Payment_gui, PhotoImage(file="LPU-logo.png")) 
    Payment_gui.title("PARKING MANAGEMENT SYSTEM")
    
    heading = Label(Payment_gui,text="LPU  CAR  PARKING  MAMAGEMENT  SYSTEM",font=('times', 35, 'italic bold underline'),bg=('yellow')) 
    
    payment = Label(Payment_gui,text="PAYMENT",font=('times', 25, 'italic bold'),bg=('yellow'))
    hours = Label(Payment_gui,text=("You used our service for approx "+str(int(hours_used))+" hours"+"\
     \n you need to pay "+ str(amount)+"Rupees"),font=('times', 25, 'italic bold'),height=5)
    
    proceed = Button(Payment_gui,text="Proceed to pay",font=('times', 25, 'italic bold'),command=Complete)
    
    logo_image.grid(row=0,column=0)
    heading.grid(row=0,column=1)
    payment.grid(row=1,column=1)
    hours.grid(row=2,column=1)
    proceed.grid(row=3,column=1,pady=20)
    
    Payment_gui.mainloop()
def Complete():
    global pay_id
    Database = connect("DataBase.db")
    current = str(datetime.now())
    temp = user[0]
    Database.execute("insert into payment values(?,?,?,?,?)",(pay_id,temp,amount,current,'Paid'))
    Database.commit()
    messagebox.showinfo("Success","payment is done\nNow you can have your vehicle")
    Database.close()
    pay_id += 1
    file = open("pay_id","w")
    file.write(str(pay_id))
    file.close()
    Payment_gui.destroy()
    MainPage()
    
def AdminLogin():
    root.destroy()
    global AdminLogin_gui
    AdminLogin_gui = Tk()
    
    AdminLogin_gui.geometry("%dx%d" % (screen_width,screen_height))
    AdminLogin_gui.config(bg="yellow")
    
    logo = PhotoImage(file="LPU-logo.png")
    logo_image = Label(AdminLogin_gui,image=logo,bg='yellow')

    AdminLogin_gui.iconphoto(AdminLogin_gui, PhotoImage(file="LPU-logo.png")) 
    AdminLogin_gui.title("PARKING MANAGEMENT SYSTEM")
    
    heading = Label(AdminLogin_gui,text="LPU  CAR  PARKING  MAMAGEMENT  SYSTEM",font=('times', 35, 'italic bold underline'),bg=('yellow')) 
    
    user_id = Label(AdminLogin_gui,text="User ID",font=('times',20,'bold'),bg='yellow')
    password = Label(AdminLogin_gui,text="Password",font=('times',20,'bold'),bg='yellow')
    
    global user_id_e,password_e
    user_id_e = Entry(AdminLogin_gui,bd=3,width=50)
    password_e = Entry(AdminLogin_gui,bd=3,width=50)
    
    confirm = Button(AdminLogin_gui,text="Confirm",font=('times',20,'bold'),command =LoginAdmin)
    back = Button(AdminLogin_gui,text="Back",font=('times',20,'bold'),command = lambda: GoBack(AdminLogin_gui))

    logo_image.grid(row=0,column=0)
    heading.grid(row=0,column=1)
    user_id.grid(row=2,column=0,pady=30)
    password.grid(row=4,column=0,pady=30)
    user_id_e.grid(row=2,column=1,pady=30)
    password_e.grid(row=4,column=1,pady=30)
    confirm.grid(row=6,column=0,pady=30,padx=80)
    back.grid(row=6,column=1,pady=30,padx=30)
    
    AdminLogin_gui.mainloop()
    
def LoginAdmin():
    user_id=user_id_e.get()
    password=password_e.get()

    if user_id=='11802503':
        if password == "himanshu":
            messagebox.showinfo("success","Login sucessfull")
            AdminLogin_gui.destroy()   
            Admin()
        else:
            messagebox.showerror("invalid","user id password does not exist")
    else:
         messagebox.showerror("invalid","user id password does not exist")
    
def Admin():
    
    global Admin_gui
    Admin_gui = Tk()
    
    Admin_gui.geometry("%dx%d" % (screen_width,screen_height))
    Admin_gui.config(bg="yellow")
    
    logo = PhotoImage(file="LPU-logo.png")
    logo_image = Label(Admin_gui,image=logo,bg='yellow')

    Admin_gui.iconphoto(Admin_gui, PhotoImage(file="LPU-logo.png")) 
    Admin_gui.title("PARKING MANAGEMENT SYSTEM")
    
    heading = Label(Admin_gui,text="ADMINISTRATOR",font=('times', 35, 'italic bold underline'),bg=('yellow')) 
    
    see_details = Button(Admin_gui,text="view database",font=('times', 25, 'italic bold'),command=View,height=1,width=12)
    back = Button(Admin_gui,text="Back",font=('times', 25, 'italic bold'),command= lambda: GoBack(Admin_gui),height=1,width=12)
    
    heading.grid(row=0,column=0,pady=30)
    see_details.grid(row=1,column=0,pady=30)
    back.grid(row=2,column=0)
    
    Admin_gui.mainloop()
    
def View():
    Admin_gui.destroy()
    global View_gui
    View_gui = Tk()
    View_gui.geometry("%dx%d" % (screen_width,screen_height))
    View_gui.config(bg="yellow")
    
    logo = PhotoImage(file="LPU-logo.png")
    logo_image = Label(View_gui,image=logo,bg='yellow')

    View_gui.iconphoto(View_gui, PhotoImage(file="LPU-logo.png")) 
    View_gui.title("PARKING MANAGEMENT SYSTEM")
    
    heading = Label(View_gui,text="ADMINISTRATOR",font=('times', 35, 'italic bold underline'),bg=('yellow')) 
    
    user_db = Button(View_gui,text="User",font=('times', 25, 'italic bold'),command= lambda: ShowTable("user"),height=1,width=12)
    park_slot = Button(View_gui,text="Parking slot",font=('times', 25, 'italic bold'),command= lambda: ShowTable("parking_slot"),height=1,width=12)
    record = Button(View_gui,text="Record",font=('times', 25, 'italic bold'),command= lambda: ShowTable("record"),height=1,width=12)
    pay = Button(View_gui,text="payment",font=('times', 25, 'italic bold'),command= lambda: ShowTable("payment"),height=1,width=12)
    back = Button(View_gui,text="Back",font=('times', 25, 'italic bold'),command= lambda: GoBack(View_gui),height=1,width=12)
    
    
    heading.grid(row=0,column=0,pady=20)
    user_db.grid(row=1,column=0,pady=20)
    park_slot.grid(row=2,column=0,pady=20)
    record.grid(row=3,column=0,pady=20)
    pay.grid(row=4,column=0,pady=20)
    back.grid(row=5,column=0,pady=20)
    View_gui.mainloop()
    
def ShowTable(name):    
    text = Text(View_gui,height=35,width=100)
    Database = connect("database.db")
    if name=="user":
        x = Database.execute("select * from user")
        x =  x.fetchall()
        text.insert(END,"(|user_id| , |password| , |name| , |date of birth| , |phone number| , |vehicle_no|)\n")
        for i in x:
            text.insert(END,i)
            text.insert(END,"\n")
    elif name=="payment":
        x = Database.execute("select * from payment")
        x =  x.fetchall()
        text.insert(END,"(|payment_id| , |user_id| , |amount| , |paid on| , |status|)\n")
        for i in x:
            text.insert(END,i)
            text.insert(END,"\n")
    elif name=="record":
        x = Database.execute("select * from record")
        x =  x.fetchall()
        text.insert(END,"(|user_id| , |time_in| , |time_out|)\n")
        for i in x:
            text.insert(END,i)
            text.insert(END,"\n")
    elif name=="parking_slot":
        x = Database.execute("select * from parking_slot")
        x =  x.fetchall()
        text.insert(END,"(|slot_id| , |user_id| , |status|)\n")
        for i in x:
            text.insert(END,i)
            text.insert(END,"\n")
            
    text.grid(row=1,rowspan=5,column=1,padx=50)
    View_gui.mainloop()
    
def GoToAdmin(gui):
    gui.destroy()
    Admin()
    
MainPage()    
```

```python
from tkinter import *
def Payment():    
    Payment_gui = Tk()
    screen_width = Payment_gui.winfo_screenwidth()
    screen_height = Payment_gui.winfo_screenheight()
    Payment_gui.geometry("%dx%d" % (screen_width,screen_height))
    Payment_gui.config(bg="yellow")
    
    logo = PhotoImage(file="LPU-logo.png")
    logo_image = Label(Payment_gui,image=logo,bg='yellow')

    Payment_gui.iconphoto(Payment_gui, PhotoImage(file="LPU-logo.png")) 
    Payment_gui.title("PARKING MANAGEMENT SYSTEM")
    
    heading = Label(Payment_gui,text="LPU  CAR  PARKING  MAMAGEMENT  SYSTEM",font=('times', 35, 'italic bold underline'),bg=('yellow')) 
    
    payment = Label(Payment_gui,text="PAYMENT",font=('times', 35, 'italic bold'),bg=('yellow'))
    hours = Label(Payment_gui,text=("You used our service for approx "+"34"+" hours"+"\
    \n you need to pay "+ str(53)+"Rupees"),font=('times', 25, 'italic bold'),height=5)
    proceed = Button(Payment_gui,text="Proceed to pay",font=('times', 25, 'italic bold'))
    logo_image.grid(row=0,column=0)
    heading.grid(row=0,column=1)
    payment.grid(row=1,column=1)
    hours.grid(row=2,column=1)
    proceed.grid(row=3,column=1,pady=20)
    
    Payment_gui.mainloop()
Payment()    
```
